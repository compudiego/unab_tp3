Definición de clase Punto

```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def eje_x(self):
        return self.x

    def eje_y(self):
        return self.y

    def impresion(self):
        return "El punto se encuentra en x = " + str(self.x) + " y en y = " + str(self.y)

    def opuesto(self):
        return Punto(-self.x, -self.y)
```

Pruebas de clase Punto,
Instancias y uso de métodos

```python
punto_a = Punto(1, 2) #instancia de un punto
print(punto_a.impresion())

punto_b = punto_a.opuesto() #instancia del punto opuesto al punto_a
print(punto_b.impresion())
```

Definición de clase Linea

```python
class Linea:
  def __init__(self, punto1, punto2):
    self.p1 = punto1
    self.p2 = punto2

  def mueveDerecha(self, n):
    self.p1.x += n
    self.p2.x += n

  def mueveIzquierda(self, n):
    self.p1.x -= n
    self.p2.x -= n

  def mueveArriba(self, n):
    self.p1.y += n
    self.p2.y += n

  def mueveAbajo(self, n):
    self.p1.y -= n
    self.p2.y -= n
```

Pruebas de clase Linea,
Instancia y prueba de métodos




```python
linea = Linea(punto_a,punto_b) #instancia de una linea a partir de los puntos creados anteriormente

print(linea.p1.impresion()) #impresion de la linea original
linea.mueveDerecha(2) #movimiento del punto hacia la derecha 2 posiciones
print(linea.p1.impresion()) #impresion de la linea luego del corrimiento
```

Definición de clase Cancion

```python
class Cancion:
  def __init__(self, titulo, autor):
    self.autor = autor
    self.titulo = titulo
  def get_autor(self):
    return self.autor
  def get_titulo(self):
    return self.titulo
  def set_titulo(self, titulo):
    self.titulo = titulo
  def set_autor(self, autor):
    self.autor = autor
```

Definición de clase Persona

```python
class Persona:
  def __init__(self, nombre, apellido, dni):
    self.nombre = nombre
    self.apellido = apellido
    self.dni = dni
  def impresion(self):
    return self.apellido + ", " + self.nombre
```

Definición de clase Autor

```python
class Autor(Persona):
    def __init__(self, nombre, apellido, dni, genero_literario):
        super().__init__(nombre, apellido, dni)
        self.genero_literario = genero_literario

    def presentarse_como_autor(self):
        return f"Soy {self.nombre} {self.apellido} y escribo {self.genero_literario}"

```

Definición de clase Lugar

```python
class Lugar:
  def __init__(self, ciudad, pais):
    self.ciudad = ciudad
    self.pais = pais
  def impresion(self):
    return self.ciudad + ", " + self.pais
```

Definición de clase Libro

```python
class Libro:
  def __init__(self, titulo, autor, ISBN, paginas, edicion, editorial, lugar, fechaEdicion):
    self.titulo = titulo
    self.autor = autor
    self.ISBN = ISBN
    self.paginas = paginas
    self.edicion = edicion
    self.editorial = editorial
    self.lugar = lugar
    self.fechaEdicion = fechaEdicion

  #Todos los getters de Libro. Estos sirven para traer informacion
  def get_titulo(self):
    return self.titulo

  def get_autor(self):
    return self.autor

  def get_ISBN(self):
    return self.ISBN

  def get_paginas(self):
    return self.paginas

  def get_edicion(self):
    return self.edicion

  def get_editorial(self):
    return self.editorial

  def get_lugar(self):
    return self.lugar

  def get_fechaEdicion(self):
    return self.fechaEdicion

  def mostrarInformacion(self):
    print("Título: " + self.titulo + " " + self.edicion +"\nAutor: " + self.autor.impresion() + "\nISBN: " + self.ISBN + "\n"+self.lugar.impresion()+ "\n"+self.fechaEdicion+ "\n"+self.paginas+ " páginas")

```

Prueba de clase Libro, instancias y uso de métodos

```python
autor = Autor("Gabriel", "Garcia Marquez", "12345678", "Variado") #instancia de un Autor
lugar1 = Lugar("Buenos Aires", "Argentina") #instancia de Lugar
libro = Libro("Cien años de soledad", autor,"9780060929794","784","2da edición","Sudamericana",lugar1,"Mayo 1967") #instancia de un nuevo Libro
libro.mostrarInformacion()
```
