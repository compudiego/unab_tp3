# 🧩 Ejemplo POO: Relación “es un” (Herencia)

**Un Perro es un Animal**

## 🏗️ Definición de clases

```python
class Animal:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def hacer_sonido(self):
        return "Hace un sonido"

    def describir(self):
        return f"{self.nombre} tiene {self.edad} años"
```

```python
class Perro(Animal):
    def __init__(self, nombre, edad, raza):
        super().__init__(nombre, edad)
        self.raza = raza

    def hacer_sonido(self):
        return "Guau"

    def traer_pelota(self):
        return f"{self.nombre} está trayendo la pelota"
```

## 🧱 Propiedades

**Animal:**
- nombre
- edad

**Perro:**
- nombre (heredada)
- edad (heredada)
- raza

## ⚙️ Métodos

**Animal:**
- hacer_sonido()
- describir()

**Perro:**
- hacer_sonido() (sobrescrito)
- traer_pelota()

## 🧪 Instanciación de objetos

```python
# Objetos de Animal
animal1 = Animal("Bicho", 7)
animal2 = Animal("Pepe", 3)

# Objetos de Perro
perro1 = Perro("Firulais", 4, "Labrador")
perro2 = Perro("Rex", 2, "Ovejero Alemán")

print(animal1.describir())
print(perro1.describir())
print(perro1.traer_pelota())
```

## 🔁 Polimorfismo

```python
print(animal1.hacer_sonido())
print(perro1.hacer_sonido())
```

## 🧠 Pilares de la POO

### 🔒 Encapsulamiento
Los datos están dentro de los objetos.

### 🎯 Abstracción
Animal representa una idea general.

### 🧬 Herencia
Perro hereda de Animal.

### 🔁 Polimorfismo
El mismo método tiene comportamientos distintos.
